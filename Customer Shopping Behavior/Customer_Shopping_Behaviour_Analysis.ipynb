```python
import pandas as pd
df = pd.read_csv(r'C:\Users\USER\Desktop\customer_shopping_behavior.csv')
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace (" ", "_")
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
df.columns
```

```python
# create a column age group
labels = ['Young Adult', 'Adults', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
df[['age', 'age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Monrhs' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days', 'frequency_of_purchases' ]].head(10)
        
```

```python
df.shape
```

```python
df.to_csv(r'C:\Users\USER\Desktop\customer_shopping_behaviour_clean_data.csv', index = False)
```
